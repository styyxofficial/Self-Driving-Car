```python
from tkinter import N
import pygame
from pygame.locals import *
import sys
import time
import math
```

```python
screen_width = 1920
screen_height = 1080
```

```python
def on_init():  
    pygame.init()


on_init()
```

```python
screen = pygame.display.set_mode((screen_width, screen_height), pygame.HWSURFACE | pygame.DOUBLEBUF)
screen.fill("WHITE")
```

```python
background_image = pygame.image.load("../images/map.png").convert_alpha()
```

```python
background_image.get_at((0, 0))
```

```python
# x, y, theta, distance
distance_data = []
```

```python
for x in range(0, screen_width, 15):
    for y in range(0, screen_height, 15):
        
        if background_image.get_at((x, y)) == (0, 0, 0, 255):
            for theta in range(0, 360, 5):
                
                if (theta > 90 and theta<270):
                    flip = -1
                else:
                    flip = 1
                    
                slope = -math.tan(math.radians(theta))
                for x1 in range(x, screen_width*flip, flip):
                    y1 = -slope * (x-x1)+y
                    
                    pos = (round(x1), round(y1))
                    try:
                        if (background_image.get_at(pos) == (255, 255, 255, 255)):
                            distance = math.sqrt((pos[0]-x)**2 + (pos[1]-y)**2)
                            distance_data.append([x, y, theta, distance])
                            break
                    except:
                        # Means the slope was infinite, and the point could not be calculated
                        continue
```

```python
import numpy as np
```

```python
np.savetxt('distance_data.csv', np.array(distance_data), delimiter=",")
```
